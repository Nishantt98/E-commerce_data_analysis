```python
import pandas as pd         #  for data cleaning 
import plotly.express as px # for visualization
import plotly.graph_objects as go  # for advance and customize graph
import plotly.io as pio  # Customize the graph template
import plotly.colors as colors  # for Color
pio.templates.default = "plotly_white"   
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")
```

#  Converting Column data

```python
data['Order Date'] = pd.to_datetime(data["Order Date"]) 
```

```python
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

# Create new Columns

```python
data["Order Month"] = data["Order Date"].dt.month 
data["Order Year"] = data["Order Date"].dt.year 
data["Order of the Week"] = data["Order Date"].dt.dayofweek
data.head(2)
```

# Monthly Sales

```python
Monthly_Sales = data.groupby("Order Month")["Sales"].sum().reset_index()
fig = px.line(Monthly_Sales,
             x='Order Month',
             y='Sales',
             title="Monthly Sales Graph")
fig.show()
```

# Sales by Category

```python
Sales_Category = data.groupby("Category")["Sales"].sum().reset_index()
```

```python
Sales_Category
```

```python
fig = px.pie(Sales_Category,
            values='Sales',
            names='Category',
            hole=0.2,
            color_discrete_sequence = px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_text="Sales Analysis by Category",title_font=dict(size=24))

fig.show()
```

# Sales By Sub-categories

```python
Sales_Sub_Category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(Sales_Sub_Category,
            x='Sub-Category',
            y='Sales',
            title="Sales Analysis by Sub category")
fig.show()
```

# Monthly Profit Analysis

```python
Monthly_profit = data.groupby("Order Month")['Profit'].sum().reset_index()

fig = px.line(Monthly_profit,
             x='Order Month',
             y='Profit',
             title='Monthly Profit Analysis')
fig.show()
```

```python
Monthly_profit = data.groupby("Order Month")['Profit'].sum().reset_index()

fig = px.bar(Monthly_profit,
             x='Order Month',
             y='Profit',
             title='Monthly Profit Analysis')
fig.show()
```

# Profit by Category

```python
Profit_Category = data.groupby("Category")["Profit"].sum().reset_index()
Profit_Category
```

# Profit By Sub Category

```python
profit_by_category = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_category,
            x='Sub-Category',
            y='Profit',
            title='Profit Analysis by Sub Category')
fig.show()
```

# Analysis Sales and Profit by Customer Segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                    y=sales_profit_by_segment['Sales'],
                    name='Sales',
                    marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                    y=sales_profit_by_segment['Profit'],
                    name='Profit',
                    marker_color=color_palette[1]))

fig.update_layout(title='Sales and profit analysis by customerr segment',
                 xaxis_title='Customer Segment',yaxis_title='Amount')

fig.show()
```

# Sales to profit Ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
